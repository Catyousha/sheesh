```python
from google_play_scraper import Sort, reviews_all

result = reviews_all(
    'com.miHoYo.GenshinImpact',
    sleep_milliseconds=0,
    lang='id',
    country='id',
    sort=Sort.MOST_RELEVANT,
    filter_score_with=5,
)
result
```

```python
import pandas as pd
data = pd.DataFrame.from_dict(result)
data
```

```python
data.to_csv('reviews.csv', index=False)
```
