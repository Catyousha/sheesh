# Case Folding

```python
caseFolded_reviews = [review.casefold() for review in reviews]

print("Sebelum Casefolding:", reviews[0], '\n')
print("Sesudah Casefolding:", caseFolded_reviews[0],'\n')
```

# Punctuation

```python
import re
removedPunc_reviews = list()
for review in caseFolded_reviews:
    clean1 = re.sub('[^a-zA-Z]', ' ', review)
    clean2 = re.sub('\s+', ' ', clean1).strip()
    removedPunc_reviews.append(clean2)
```

```python
print("Sebelum Punctuation Removal:", caseFolded_reviews[0], '\n')
print("Sesudah Punctuation Removal:", removedPunc_reviews[0],'\n')
```

# Slang Removal

```python
kamus_alay = pd.read_csv("kamusalay.csv")
kamus_alay.head(3)
```

```python
alay_dict = dict(zip(kamus_alay.slang, kamus_alay.formal))
```

```python
removedSlangs_reviews = list()
for review in removedPunc_reviews:
    newReview = list()
    for word in review.split():
        if word in alay_dict.keys():
            newReview.append(alay_dict[word])
        else:
            newReview.append(word)
    removedSlangs_reviews.append(' '.join(newReview))
```

```python
idx = 0
print("Sebelum slang removal: ", removedPunc_reviews[idx], '\n')
print("Sesudah slang removal: ", removedSlangs_reviews[idx], '\n')
```
