# Case Folding

```python
caseFolded_reviews = list()
for review in reviews:
    caseFolded_reviews.append(review.casefold())

print("Sebelum Casefolding:", reviews[0], '\n')
print("Sesudah Casefolding:", caseFolded_reviews[0],'\n')
```

# Punctuation

```python
import re
removedPunc_reviews = list()
for review in caseFolded_reviews[0:5]:
    words = list()
    for word in review:
        clean1 = re.sub('[^a-zA-Z\s]', ' ', word)
        # WIP: EXTRA WHITESPACES
        clean2 = re.sub('\s+', '=', clean1)
        words.append('  '.join(clean2))
    removedPunc_reviews.append(''.join(words))

removedPunc_reviews
```
