```python
import pandas as pd
reviews = pd.read_csv("../01_preprocess/result.csv", dtype="string").dropna()
reviews
```

# Tokenizing

```python
from nltk.tokenize import word_tokenize


tokenized_reviews = [word_tokenize(review) for review in reviews.reviews]
tokenized_reviews
```

# Dictionarization

```python
import gensim as gsm

dictionary = gsm.corpora.Dictionary(tokenized_reviews)

dictionary.filter_extremes(no_below=10, no_above=0.5, keep_n=100000)
for k, v in dictionary.iteritems():
    print(k, v)
```

# Bag of Words

```python
bow_corpus = [dictionary.doc2bow(doc) for doc in tokenized_reviews]
bow_corpus_sample = bow_corpus[42]
for i in range(len(bow_corpus_sample)):
    print(f"Kata ('{dictionary[bow_corpus_sample[i][0]]}') muncul {bow_corpus_sample[i][1]} kali di review ini.")
    
```

# TF-IDF

```python
tfidf = gsm.models.TfidfModel(bow_corpus)
corpus_tfidf = tfidf[bow_corpus]
```

```python
from pprint import pprint

for doc in corpus_tfidf:
    pprint(doc)
    break
```

# Running LDA

```python
NUM_TOPICS = 5
lda_model = gsm.models.LdaMulticore(bow_corpus, num_topics=NUM_TOPICS, id2word=dictionary, passes=2, workers=2)
```

```python
for idx, topic in lda_model.print_topics(-1):
    print('Topic: {} \nWords: {}'.format(idx, topic))
```

```python
lda_model_tfidf = gsm.models.LdaMulticore(corpus_tfidf, num_topics=NUM_TOPICS, id2word=dictionary, passes=2, workers=2)
```

```python
for idx, topic in lda_model_tfidf.print_topics(-1):
    print('Topic: {} \nWords: {}'.format(idx, topic))
```

```python
coherence_model_lda = gsm.models.CoherenceModel(model=lda_model, texts=tokenized_reviews, dictionary=dictionary, coherence='c_v')

coherence_model_lda.get_coherence()
```
